```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
####### ------- RMSE values ------- #######

RMSE_values = {
    'MLR': (36.47, 43.53, 42.05, 37.98, 32.69, 44.25, 37.37),
    'RM F':	(29.69, 30.14, 25.49, 24.28, 21.61, 26.48, 28.16),
    'RM B':	(31.7, 30.14, 25.49, 24.65, 22.65, 27.68, 29.63),
    'RM S': (29.69, 30.14, 24.42, 21.38, 20.53, 27.37, 29.63),
    'PCR':	(26.93, 25.38, 21.5, 19.47, 20.92, 21.37, 22.78),
    'PLS': (28.01, 25.92, 22.43, 21.16, 21.53, 22.11, 22.13),
    'RF': (26, 25.49, 23.83, 22.75, 21.21, 21.94, 22.85)
}

lags = ("Lag0", "Lag1", "Lag2", "Lag3", "Lag4", "Lag5", "Lag6")
x = np.arange(len(lags))
w = 0.1
m = 0

fig, ax = plt.subplots(layout='constrained')
for model, values in RMSE_values.items():
    offset = w * m
    rects = ax.bar(x + offset, values, w, label=model)
    m += 1

ax.set_ylabel('RMSE')
ax.set_xticks(x + w * len(RMSE_values) / 2, lags)
ax.legend(loc='upper left', ncols=7, fontsize=7)
ax.set_ylim(15, 50)
plt.title('RMSE Using LBar Graph')
plt.show()
```
