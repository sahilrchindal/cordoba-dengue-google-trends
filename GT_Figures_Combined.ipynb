```python
# Import packages
# import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
####### ------- RMSE values ------- #######

RMSE_values = {
    'MLR': (44.21, 55.55, 49.96, 40.69, 33.67, 53.94, 39.9),
    'RM F':	(35.5, 29.16, 32.17, 26.16, 20.5, 25.64, 26.49),
    'RM B':	(36.51, 27.71, 32.17, 23.71, 20.5, 25.07, 26.68),
    'RM S': (35.5, 26.09, 32.17, 26.38, 21.86, 25.64, 26.34),
    'PCR':	(28.41, 28.04, 25.75, 21.49, 23.13, 23.21, 23.99),
    'PLS': (29.26, 27.75, 26.79, 22.07, 23.02, 23.48, 24.07),
    'RF': (27.99, 27.23, 27.01, 22.78, 23.78, 24.17, 23.98)
}

lags = ("Lag0", "Lag1", "Lag2", "Lag3", "Lag4", "Lag5", "Lag6" )
x = np.arange(len(lags))
w2 = 0.1
m2 = 0
fig, ax = plt.subplots(layout='constrained')
for model, values in RMSE_values.items():
    offset = w2 * m2
    rects = ax.bar(x + offset, values, w2, label=model)
    m2 += 1

ax.set_ylabel('RMSE')
ax.set_xticks(x + w2 * len(RMSE_values) / 2, lags)
ax.legend(loc='upper left', ncols=7, fontsize=7)
ax.set_ylim(15, 60)
plt.title('RMSE Using Bar Graph')
plt.show()
```
